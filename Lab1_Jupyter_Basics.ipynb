1. **What are some differences between markdown, text, and code cells?**  
   - Markdown cells are used for documentation, notes, and formatting text.  
   - Text cells (essentially Markdown cells) are for plain or formatted explanations.  
   - Code cells are used to write and execute Python (or other language) code.

2. **When might you need to restart the kernel?**  
   - When variables need to be reset, memory usage is high, or the notebook becomes unresponsive.

3. **How could keyboard shortcuts improve your Jupyter workflow?**
